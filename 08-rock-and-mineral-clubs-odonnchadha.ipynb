```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = "https://rocktumbler.com/blog/rock-and-mineral-clubs/"
raw_html = requests.get(url).content
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(type(soup_doc))
```

```python
### SUCCESSFUL MINIMUM CODE
all_rows = []
all_tables = soup_doc.find_all('table', limit=100)
for mytable in all_tables:
    rows = mytable.find_all('tr')
    for row in rows[1:]:
        myrow = []
        info = row.find_all('td')
        for data in info:
            myrow.append(data.string)
        link = row.a['href']
        myrow.append(link)
        all_rows.append(myrow)
all_rows
```

```python
### SUCCESSFUL STATE-STRING CODE
state = 'NOSTATE'
all_rows = []
all_tables = soup_doc.find_all('section', limit=52)
for mytable in all_tables:
    if mytable.find('h3'):
        state = mytable.find('h3').string
    else:
        pass
    rows = mytable.find_all('tr')
    for row in rows[1:]:
        myrow = []
        info = row.find_all('td')
        for data in info:
            myrow.append(data.string)
        link = row.a['href']
        myrow.append(state)
        myrow.append(link)
        all_rows.append(myrow)
all_rows
# type(all_rows[1][2])
```
